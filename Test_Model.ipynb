```python
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os
from math import trunc
```

```python
Categories = ["NORMAL", "PNEUMONIA"]
```

```python
def prepare_CNN(image):
    img_size=100
    img=load_img(image, color_mode='grayscale', target_size=(img_size,img_size))
    new_array=img_to_array(img)

    return new_array.reshape(-1,img_size, img_size,1)
```

```python
model = load_model('Model/Pneumonia_Detection_CNN_87.0.h5')
```

```python
def classify_CNN(img):
    predict = model.predict([prepare_CNN(img)/255.0])
    confidence = predict[0][0] * 100
    result = Categories[int(round(predict[0][0]))]
    print("-------CNN-----")
    print(f'Image: {img}')
    print(f'Confidence: {round(confidence, 2)}%')
    print(f'Prediction: {result}')  
    return result
```
